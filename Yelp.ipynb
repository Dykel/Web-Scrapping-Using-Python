# Yelp 

```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
h = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36'
}
```

```python
html = requests.get('https://www.yelp.com/search?find_desc=Plumbers&find_loc=San+Francisco%2C+CA&ns=1',headers=h)
html.status_code
```

```python
bsobj = soup(html.content,'lxml')
bsobj
```

```python
names = []
for name in bsobj.findAll('div',{'class':'lemon--div__373c0__1mboc mainAttributes__373c0__1r0QA arrange-unit__373c0__o3tjT arrange-unit-fill__373c0__3Sfw1 border-color--default__373c0__3-ifU'})[10:30]:
    names.append(name.a.text)
    
names 
```

```python
ratings = []

for rating in bsobj.findAll('span',{'class':'lemon--span__373c0__3997G text__373c0__2Kxyz reviewCount__373c0__2r4xT text-color--black-extra-light__373c0__2OyzO text-align--left__373c0__2XGa-'})[10:30]:
    ratings.append(int(rating.text))
    
ratings
```

```python
len(ratings)
```

```python
stars = []
for star in bsobj.findAll('span',{'class':'lemon--span__373c0__3997G display--inline__373c0__3JqBP border-color--default__373c0__3-ifU'}):
    try:
        stars.append(star.div['aria-label'])
        
    except:
        continue
print(len(stars))

print(stars)
len(stars)
```
